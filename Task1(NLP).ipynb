1.**What is the difference between "Love" and "love" in NLP?**

A. "Love" and "love" differ by capitalization
(capital L vs small l).
In NLP, they are treated as two different tokens because many models are case-sensitive.
"Love" is usually used at the beginning of a sentence or as a proper noun.
"love" is a common word used to express emotion or action.
To remove this difference, NLP uses a preprocessing step called lowercasing.
After lowercasing, both "Love" and "love" become the same word.

2.**What happens if stopwords are not removed?**

A. The text will contain many common words like is, the, and, of which do not add much meaning.
These extra words create noise and make it harder for the model to focus on important words.
The size of the data increases, leading to more memory usage and slower processing.
It may reduce model accuracy, because important keywords get less importance.
The model may give less meaningful results, especially in tasks like search or classification.
However, in some cases (like sentiment analysis), keeping stopwords can be useful for context.

3.**Provide two real-world scenarios where removing stopwords can be harmful.**

A. i. Machine Translation
Sentence: "He is in the house"
If stopwords are removed: "He house"
Words like "is" and "in" are important for grammar and correct meaning. Removing them can lead to incorrect or incomplete translations.
ii. Speech Recognition / Text Understanding
Sentence: "Turn off the light"
If stopwords are removed: "Turn off light"
While still understandable, removing words like "the" may sometimes reduce clarity or natural understanding in complex commands.

4.**Explain the difference between stemming and lemmatization**.

A.*Stemming*  

It is a process of cutting off prefixes or suffixes from a word.
It uses simple rules, not understanding the actual meaning.
The result may not be a real word.
Example:
running → run
studies → studi (not a correct word)
It is faster but less accurate.

*Lemmatization*

It converts a word to its base or dictionary form (lemma).
It uses vocabulary and grammar rules.
The result is always a meaningful, valid word.
Example:
running → run
studies → study
It is slower but more accurate.

```python
#Task-2 Build Advanced Preprocessing Function
import re

def preprocess_text(text):
    # 1. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 2. Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Convert to lowercase
    text = text.lower()

    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove very short tokens (≤2) except "no", "not"
    tokens = text.split()
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in {"no", "not"}
    ]

    return " ".join(filtered_tokens)

user_input = input("Enter text: ")
result = preprocess_text(user_input)
print("Processed text:", result)
```

```python
#Task 5: Frequency Analysis
from collections import Counter

#Combine tokens
all_tokens = []
for sentence in sentences:
    tokens = sentence.lower().split()
    all_tokens.extend(tokens)

#Frequency count
freq = Counter(all_tokens)

#Top 10 most frequent words
top_10 = freq.most_common(10)

#Top 5 least frequent words
least_5 = sorted(freq.items(), key=lambda x: x[1])[:5]

#Display
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(word, ":", count)
```

```python
#Task 6: Build Full Pipeline
import re

def preprocess_text(text):
    # 1. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 2. Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Convert to lowercase
    text = text.lower()

    # 5. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove short tokens (≤2) except "no", "not"
    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in {"no", "not"}]

    return tokens, " ".join(tokens)


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Visit https://openai.com now!"
]

result = full_pipeline(sentences)

print(result)
```
